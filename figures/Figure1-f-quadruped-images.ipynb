```python
import matplotlib.pyplot as plt
import glob
import os
import sys
import matplotlib.image as mpimg
```

```python
quadruped_iid_files = glob.glob('Figure1_quadruped/iid/*.jpg')
quadruped_iid_files = sorted(quadruped_iid_files)

quadruped_pred_iid_files = glob.glob('pred_quadruped_test_iid/*.jpg')
quadruped_pred_iid_files = sorted(quadruped_pred_iid_files)


quadruped_ood_files = glob.glob('Figure1_quadruped/ood/*.jpg')
quadruped_ood_files = sorted(quadruped_ood_files)


quadruped_pred_ood_files = glob.glob('pred_quadruped_test_ood/*.jpg')
quadruped_pred_ood_files = sorted(quadruped_pred_ood_files)
```

```python
fig, axes = plt.subplots(4,4)

for idx in range(16):
    row = idx //4
    col = idx % 4
    
    if idx > 12:
        axes[row][col].axis('off')
        continue    
    
    file = quadruped_iid_files[idx]
    img =  mpimg.imread(file)    
    axes[row][col].imshow(img)
    axes[row][col].set_xticks([])
    axes[row][col].set_yticks([])                

plt.subplots_adjust(left=0.05, right=0.95, bottom=0.05, top=0.95, wspace=0.05, hspace=0.05)
fig.set_size_inches(20, 12) 
plt.show()
```
